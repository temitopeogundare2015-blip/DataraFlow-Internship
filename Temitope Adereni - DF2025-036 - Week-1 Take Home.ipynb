```python
#program that prints the multiplication table of a number (from 1‚Äì12)
number = int(input("Enter a number: "))

print(f"\nMultiplication Table of {number}:\n")
for i in range(1, 12):
    print(f"{number} x {i} = {number * i}")
```

```python
#program that calculates the factorial of a number using a loop
number = int(input("Enter a number: "))
if number < 0:
    print("Sorry, factorial does not exist for negative numbers.")
elif number == 0 or number == 1:
    print(f"The factorial of {number} is 1")
else:
    factorial = 1
    for i in range(1, number + 1):
        factorial *= i
    print(f"The factorial of {number} is {factorial}")
```

```python
#program that finds the largest number in a given list
number = [5, 10, 15, 20, 25, 30]
print(max(number))
```

```python
#program that counts how many vowels are in a given string
text = input("Enter a string: ")
vowels = "aeiouAEIOU"
count = sum(1 for char in text if char in vowels)
print("Number of vowels:", count)
```

```python
#function that takes two numbers and returns their sum, difference, and product
def calculate_operations(a, b):
    sum = a + b
    difference = a - b
    product = a * b
    return [sum, difference, product]
a = 25
b = 5
result = calculate_operations(a, b)

print(f"Sum: {result[0]}")
print(f"Difference: {result[1]}")
print(f"Product: {result[2]}")
```

```python
#program that reverses a string without using slicing (`[::-1]`)
def reverse_string(text):
    reversed_text = ""
    for char in text:
        reversed_text = char + reversed_text 
    return reversed_text
text = input("enter a string: ")
print("reversed string:", reverse_string(text))    
```

```python
#program that generates the first 10 numbers of the Fibonacci sequence
a, b = 0, 1
print("First 10 Fibonacci numbers:")
for _ in range(10):
    print(a, end=" ")
    a, b = b, a + b
```

```python
#Shopping Cart 

while True:
    print("\n--- Shopping Cart Menu ---")
    print("1. Add item")
    print("2. View cart")
    print("3. Drop item")
    print("4. Checkout")
    
    choice = input("Enter your choice (1/2/3/4): ")
    
    if choice == "1":
        item = input("Enter item name: ")
        price = float(input("Enter item price: "))
        shopping_cart[item] = price
        print(f"{item} added to cart for ${price:.2f}")
        
    elif choice == "2":
        if not shopping_cart:
            print("Your cart is empty.")
        else:
            print("\nItems in your cart:")
            for item, price in shopping_cart.items():
                print(f"- {item}: ${price:.2f}")
    elif choice == "3":
        if not shopping_cart:
            print("Your cart is empty, no item to drop")
        else:
            item_to_remove = input("Enter the name of the item to drop: ")
            if item_to_remove in shopping_cart:
                if shopping_cart[item_to_remove][1] > 1:
                    shopping_cart[item_to_remove][1] -= 1
                    print(f"One {item_to_remove} removed from cart.")
                else:
                    del shopping_cart[item_to_remove]
                    print(f"{item_to_remove} removed from cart.")
            else:
                print(f"{item_to_remove} is not in the cart.")
    elif choice == "4":
        total = sum(shopping_cart.values())
        print("\n--- Checkout ---")
        if not shopping_cart:
            print("Your cart is empty.")
        else:
            for item, price in shopping_cart.items():
                print(f"- {item}: ${price:.2f}")
            print(f"Total cost: ${total:.2f}")
        print("Thank you for shopping!")
        break
    else:
        print("Invalid choice, please try again.")
```

```python
#File Handling
student_names = []
num_students = int(input("How many student names do you want to enter? "))
for i in range(num_students):
    name = input(f"Enter name of student {i+1}: ")
    student_names.append(name)
with open("students.txt", "w") as file:
    for name in student_names:
        file.write(name + "\n")   
print("\nStudent names saved to 'students.txt'.")
print("\nReading names from file:")
with open("students.txt", "r") as file:
    for line in file:
        print(line.strip())
```
