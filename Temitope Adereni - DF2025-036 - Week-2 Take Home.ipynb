Create an iterator class `EvenNumbers` that takes a limit and iterates over even numbers only. Use it in a `for` loop.

```python
class EvenNumbers:
    def __init__(self, limit):
        self.limit = limit
        self.current = 0
    
    def __iter__(self):
        return self
    
    def __next__(self):
        if self.current >= self.limit:
            raise StopIteration
        result = self.current
        self.current += 2
        return result
print("Even numbers up to 10:")
for num in EvenNumbers(10):
    print(num, end=" ")
```

Write a program demonstrating:

- a global variable modified inside a function

- a nonlocal variable modified inside a nested function.


```python
counter = 0
def outer_function():
    outer_counter = 0
    
    def inner_function():
        global counter
        counter += 1
        
        nonlocal outer_counter
        outer_counter += 1
        
        print(f"Inside inner_function:")
        print(f"  Global counter: {counter}")
        print(f"  Outer counter: {outer_counter}")
    
    print("First call to inner_function:")
    inner_function()
    
    print("\nSecond call to inner_function:")
    inner_function()
    
    print(f"\nAfter calls, in outer_function:")
    print(f"  Global counter: {counter}")
    print(f"  Outer counter: {outer_counter}")

outer_function()

print(f"\nIn global scope:")
print(f"  Global counter: {counter}")
```

Create a class `Fibonacci` that generates Fibonacci numbers up to a given `n` using an iterator protocol.

```python
class Fibonacci:
    def __init__(self, n):
        self.n = n
        self.count = 0
        self.a, self.b = 0, 1
    
    def __iter__(self):
        self.count = 0
        self.a, self.b = 0, 1
        return self
    
    def __next__(self):
        if self.count >= self.n:
            raise StopIteration
        
        if self.count == 0:
            self.count += 1
            return 0
        elif self.count == 1:
            self.count += 1
            return 1
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.b

print("First 10 Fibonacci numbers:")
for num in Fibonacci(10):
    print(num, end=" ")
```

Create classes `Rectangle`, `Circle`, and `Triangle`, each with an `area()` method.  
- Store instances of these classes in a list.  
- Write code that calculates the total area without checking object types.  

```python
import math

class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height
    
    def area(self):
        return self.width * self.height
    
    def __str__(self):
        return f"Rectangle(width={self.width}, height={self.height})"

class Circle:
    def __init__(self, radius):
        self.radius = radius
    
    def area(self):
        return math.pi * self.radius ** 2
    
    def __str__(self):
        return f"Circle(radius={self.radius})"

class Triangle:
    def __init__(self, base, height):
        self.base = base
        self.height = height
    
    def area(self):
        return 0.5 * self.base * self.height
    
    def __str__(self):
        return f"Triangle(base={self.base}, height={self.height})"

shapes = [
    Rectangle(6, 4),
    Circle(4),
    Triangle(5, 8),
    Rectangle(2, 9),
    Circle(6)
]

total_area = 0
for shape in shapes:
    total_area += shape.area()

print("Individual shape areas:")
for shape in shapes:
    print(f"{shape}: {shape.area():.2f}")

print(f"\nTotal area of all shapes: {total_area:.2f}")
```
