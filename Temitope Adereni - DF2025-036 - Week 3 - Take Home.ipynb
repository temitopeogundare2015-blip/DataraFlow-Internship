```python
#Question 1
import json
students = [
     { "name":"John", "age":30, "grade":"A" },
     { "name":"Tobi", "age":25, "grade":"C" },
     { "name":"Tosin", "age":32, "grade":"B" }
     ]

with open("students.json", "w") as file:
    json.dump(students, file, indent=4)

print("Student data saved to students.json")

with open("students.json", "r") as file:
    saved_records = json.load(file)
    print("\nsaved_records:")
for student in saved_records:
    print(student)
```

```python
#Question 2
import random

score = 0
operations = {
    '+': lambda x, y: x + y,
    '-': lambda x, y: x - y,
    '*': lambda x, y: x * y,
    '/': lambda x, y: x / y
}

for i in range(5):
    num1 = random.randint(1, 20)
    num2 = random.randint(1, 20)
    op = random.choice(list(operations.keys()))

    if op == "/":
        num1 = num1 * num2  

    answer = operations[op](num1, num2)

    user_answer = float(input(f"Question {i+1}: {num1} {op} {num2} = "))

    if abs(user_answer - answer) < 0.001:  
        print("✓ Correct!")
        score += 1
    else:
        print(f"❌ Wrong! The correct answer is {answer}")

print(f"\nYour final score is {score}/5")
```

```python
#Question 3
try:
    number = int(input("Enter a number: "))
    answer = 100 / number
    print(answer)
except ValueError:
    print("Invalid input! Enter a valid number.")
except ZeroDivisionError:
    print("You cannot divide by zero.")
```

```python
#Question 4
birthday_str = input("What's your birthday? (YYYY-MM-DD): ")
birthday = datetime.strptime(birthday_str, "%Y-%m-%d").date()

today = date.today()
current_year = today.year

next_birthday = birthday.replace(year=current_year)

if next_birthday < today:
    next_birthday = birthday.replace(year=current_year + 1)

days_left = (next_birthday - today).days

if days_left == 0:
        print("Today is your birthday!")
else:
    print(f"{days_left} days till your next birthday!")
```
