````python
#2
arr = np.array([1.4, 2+3j, 5.5, 4j, 7])

print(arr)

print("\nComplex?")
print(arr)
print(np.iscomplex(arr))

print("\nReal Numbers?")
print(arr)
print(np.isreal(arr))

print("\nScaler?")
a = 10
print(np.isscalar(a))
````

````python
#3
arr1 = np.array([5, 10.0, 20.00001, 40.0])
arr2 = np.array([5, 10.00004, 20.00002, 39.9999])

result = np.allclose(arr1, arr2)
print(result)
````

````python
#4
arr1 = np.array([1, 3, 5, 7])
arr2 = np.array([2, 4, 6, 8])

print("\nGreater than (arr1 > arr2):")
print(np.greater(arr1, arr2))

print("\nGreater equal (arr1 >= arr2):")
print(np.greater_equal(arr1, arr2))

print("\nLess than (arr1 < arr2):")
print(np.less(arr1, arr2))

print("\nLess equal (arr1 <= arr2):")
print(np.less_equal(arr1, arr2))
````

````python
#5
arr1 = np.array([10.0, 20.0, 30.0, 40.0001])
arr2 = np.array([10.0, 20.0001, 29.9999, 40.0])

# Element‑wise equality check
print(np.equal(arr1, arr2))

# Element‑wise equality within a tolerance
print(np.isclose(arr1, arr2))
````
