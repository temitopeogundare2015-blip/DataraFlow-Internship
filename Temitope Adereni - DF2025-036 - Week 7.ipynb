```python
from datetime import datetime
import matplotlib.pyplot as plt

london.index = pd.to_datetime(london.index)

spring_humidity = london['Mean Humidity'].loc[datetime(2014,3,1) : datetime(2014,5,31)]

monthly_humidity = spring_humidity.resample('ME').mean()

months = ['March', 'April', 'May']

plt.figure(figsize=(10, 6))
plt.title('Mean Humidity during Spring Season (2014)')
plt.ylabel('Mean Humidity')
plt.xlabel('Month')
plt.xticks(range(len(months)), months)
plt.bar(range(len(months)), monthly_humidity.values)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

In London, the spring season lasts from March to May every year. However, in 2014, the mean humidity in March and April was 71%. On the other hand, May had lower humidity at 69%, which was 2% lower than in March and April.

## Step-2 Finding a summer break

According to meteorologists, summer extends for the whole months of June, July, and August in the northern hemisphere and the whole months of December, January, and February in the southern hemisphere. So create a dataframe that holds just those months using the `datetime` index.

```python
import pandas as pd

LHR = pd.read_csv('London Heathrow Airport.csv')
LHR.columns = LHR.columns.str.strip()
LHR['date'] = pd.to_datetime(LHR['date'])
LHR = LHR.set_index('date')

summer_months = [6, 7, 8]
LHR_summer = LHR[LHR.index.month.isin(summer_months)].copy()

print(LHR_summer)
```

London Heathrow Airport(LHR) lies in the Northern Hemisphere; therefore, the summer months in LHR are June, July and August. This table shows the weather observations in these months in 2023. The hottest day was 31.3 degrees Celsius, while the recorded minimum temperature was 7.4 degrees Celsius during these summer months. Throughout this period, the weather was fair with moderate wind and there was no snow in sight at all.
