```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

headers = {
    'User-Agent': 'MyDataProject/1.0 (temitopeogundare2015@gmail.com)'
}

url = 'https://en.wikipedia.org/wiki/List_of_highest_mountains_on_Earth'
response = requests.get(url, headers=headers)

soup = BeautifulSoup(response.content, 'html.parser')

tables = soup.find_all('table', {'class': 'wikitable'})

mountains = pd.read_html(str(tables[0]))[0]
print(mountains)
```

```python
mountains.head(10)
```

```python
mountains.tail(10)
```

```python
top_ten = mountains[['Mountain name(s)', 'Country/ Region administered by[a]']]
top_ten.head(10)
```

```python
import matplotlib.pyplot as plt

top_10_for_scatter = top_ten.head(10)

plt.figure(figsize=(12, 8))
plt.scatter(x=top_10_for_scatter['Country/ Region administered by[a]'], y=top_10_for_scatter['Mountain name(s)'])

plt.title('Top 10 Mountain Names by Administering Country/Region')
plt.xlabel('Country/ Region administered by[a]')
plt.ylabel('Mountain Name(s)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

**Tourism and Product Development**

The world's Top 10 highest peaks are more than just tourism destinations, they are extreme product testing laboratories. Analysis of their height and prominence segments the market into distinct challenges. Peaks like K2 demand advanced, high-fatality-rate gear due to their technical difficulty, which creates a premium market for specialised mountaineering equipment. On the other hand, the more frequented, lower-prominence routes like the Everest's massif drive demand for durable, mass-market expedition gear and base camp facilities.

**Getting the Data**

Upon downloading the dataset, minor cleaning was performed by removing the empty column and grouping the dataset by country and mountain name. This is to see where the top ten highest mountains are located and to observe similarities among them.

**Showing the data**

The plot shows that three of the top ten highest mountains in the world are located in Nepal. These three mountains are Annapurna I, Manaslu and Dhaulagiri I. Four mountains are located between Nepal and China, including Mount Everest, Lhotse, Makalu and Cho Oyu. Nanga Parbat is the only mountain entirely located in Pakistan, and K2, the second-highest mountain in the world, is located between Pakistan and China.

**Conclusions**

In conclusion, aggressive marketing for mountain gear would be beneficial for sales in China and Nepal, most especially. These countries are the most visited in terms of tourism due to the presence of these tall mountains that continue to attract tourists. Marketing should be tailored towards the Spring (April-May) and Autumn (Late Sept-Nov).

However, to test our new product, Nepal and China would be our marketing hotspots.
